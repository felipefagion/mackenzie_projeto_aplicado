```python
#Manipula√ß√£o de dados
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)

#Visualiza√ß√£o de Dados
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('base_imoveis.xlsx')
```

```python
# Exibe as primeiras linhas do dataset.
df.head()
```

```python
# Informa sobre a quantidade de linhas e colunas do dataset.
print(f"O dataset possui {df.shape[0]} linhas e {df.shape[1]} colunas.")
```

```python
# Exibe o nome das colunas do dataset.
df.columns
```

```python
# Exibe a quantidade de valores unicos para cada variavel.
df.nunique()
```

```python
# Exibe informacoes sobre as variaveis.
df.info()
```

```python
# Exibe as estatisticas descritivas das variaveis numericas.
df.describe().round(2)
```

```python
# Exibe o somatorio da quantidade de dados ausentes.
df.isnull().sum()
```

```python
# Exibe os dados ausentes em porcentagem.
print("Porcentagem de dados faltantes:")
(df.isnull().sum() / len(df)* 100).apply(lambda x: f'{x:.4f}%')
```

```python
# Selecao das variaveis a serem exibidas
var = ['area_util', 'quartos', 'vagas', 'condominio', 'banheiros', 'suites', 
        'piscina', 'academia', 'quadra', 'media_bairro', 'media_idh', 
        'media_gini', 'expectativa_vida', 'renda_percapita']


# Configuracao dos subplots
n_variaveis = len(var)
n_linhas = (n_variaveis + 2) // 3

# Configuracao do tamanho dos subplots
fig, axs = plt.subplots(n_linhas, 3, figsize=(18, n_linhas * 5))


# Ajuste e configuracao do visual dos subplots e grafico de regressao.
for i, var in enumerate(var):
    linha = i // 3
    coluna = i % 3
    sns.regplot(x=var, y='valor_total', data=df, ax=axs[linha, coluna], scatter_kws={'s': 10})
    axs[linha, coluna].set_title(f'Correla√ß√£o entre valor_total e {var}')
    axs[linha, coluna].set_ylabel('Valor Total')
    axs[linha, coluna].set_xlabel(var)

# Remove subplots vazios
for j in range(i+1, n_linhas * 3):
    fig.delaxes(axs.flatten()[j])

plt.tight_layout()
plt.show()
```

```python
# Selecao das variaveis a serem exibidas
var = ['quartos', 'vagas', 'banheiros', 'suites', 
        'piscina', 'academia', 'quadra', ]


# Configuracao dos subplots
n_variaveis = len(var)
n_linhas = (n_variaveis + 2) // 3

# Configuracao do tamanho dos subplots
fig, axs = plt.subplots(n_linhas, 3, figsize=(18, n_linhas * 5))

# Ajuste e configuracao do visual dos subplots e Boxplot.
for i, var in enumerate(var):
    linha = i // 3
    coluna = i % 3
    sns.boxplot(x=var, y='valor_total', data=df, ax=axs[linha, coluna])
    axs[linha, coluna].set_title(f'Boxplot: valor_total e {var}')
    axs[linha, coluna].set_ylabel('Valor Total')
    axs[linha, coluna].set_xlabel(var)

# Remove subplots vazios
for j in range(i+1, n_linhas * 3):
    fig.delaxes(axs.flatten()[j])

plt.tight_layout()
plt.show()
```

```python
# Selecao das variaveis a serem exibidas
var2 = ['area_util', 'condominio', 'media_bairro', 'media_idh', 
        'media_gini', 'expectativa_vida', 'renda_percapita']

# Configuracao dos subplots
n_variaveis2 = len(var2)
n_linhas2 = (n_variaveis2 + 2) // 3

# Configuracao do tamanho dos subplots
fig, axs = plt.subplots(n_linhas2, 3, figsize=(18, n_linhas2 * 5))

# Ajuste e configuracao do visual dos subplots e histograma.
for i, var2 in enumerate(var2):
    linha = i // 3
    coluna = i % 3
    sns.histplot(x=var2, data=df, ax=axs[linha, coluna], kde=True)
    axs[linha, coluna].set_title(f'Histograma da variavel {var2}')
    axs[linha, coluna].set_xlabel(var2)

# Remove subplots vazios
for j in range(i+1, n_linhas2 * 3):
    fig.delaxes(axs.flatten()[j])

plt.tight_layout()
plt.show()
```
